```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

# Sample color dictionary (replace with your actual dictionary)
color_dict = {1: 'red', 2: 'green', 3: 'blue'}

# Create the scatter plot
plt.scatter(x, y)

# Extract unique colors from the dictionary
colors = list(color_dict.values())

# Extract labels from the dictionary based on the order of colors
labels = [color_dict[key] for key in color_dict.keys() if color_dict[key] in colors]

# Create the legend
plt.legend(colors, labels)

# Customize the plot (optional)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot with Color Labels")

# Show the plot
plt.show()
```
